```
import numpy as np
import seaborn as sns
from scipy.stats import *
import math
import statistics
```

```
# Functions to calculate test statistic

class find_test_statistic:
    def proportion(self, sample, hypothesis, size):
        return (sample - hypothesis)/math.sqrt(hypothesis*(1-hypothesis)/size)

    def mean(self, sample, hypothesis, size, std_dev):
        return (sample - hypothesis)/(std_dev/math.sqrt(size))

    def variance(self, sample_var, hypothesis_var, size):
        return (size - 1)*sample_var/hypothesis_var
```

```
# calculation of variance

variance = lambda x : statistics.variance(x)
```

```
find_test_statistic = find_test_statistic()
```

```
# Writing a function to print conclusion

def conclusion(p):
    significance_level = 0.05
    if p > significance_level:
        return f"Insufficient evidence, failed to reject the null hypothesis for p = {p}."
    else:
        return f"We have sufficient evidence to reject the null hypothesis for p = {p}."
```

```
# z CDF calculation

z_cdf = lambda x : norm(0,1).cdf(x)
```

```
# Calculation of p value (probability value)

def p_value(z, tailed):
    z = z_cdf(z)
    if tailed == "left":
        return z
    elif tailed == "right":
        return 1-z
    elif tailed == "two":
        if z > 0.5:
            return 2*(1-z)
        else:
            return 2*z
    else:
        return "Invalid taildness"
```

# <b><u> Problem 1 </u> </b>

## <b>The average IQ of a sample of 50 university students was found to be 105. Carry out a statistical test to determine whether the average IQ of university students is greater than 100, assuming that IQs are normally distributed. It is known from previous studies that the standard deviation of IQs among students is approximately 20. </b>

Given values : <br>


*   Sample mean = 105
*   Sample std_dev = 20
*   Sample size = 50
*   Hypothesis mean = 100
*   p > 100
*   Right tailed






1.   Null hypothesis: p = 100
2.   Alternate hypothesis: p > 100
3.   Critical value: alpha = 0.05 :: p < 0.05
4.   Identify p value
5.   Conclusion





```
hypothesis = 100
sample_mean = 105
std_dev = 20
size = 50

z = find_test_statistic.mean(sample_mean, hypothesis, size, std_dev)
print(z)
```

```
p = p_value(z, "right")   # Right Tailed
print(p)
```

```
print(conclusion(p))
```

## <b>In a one-year mortality investigation, 45 of the 250 ninety-year-olds present at the start of the investigation died before the end of the year. *   theta_hypothesis = 0.2

1. Null hypothesis: theta = 0.2

```
theta_hypothesis = 0.2
theta_population = 45/250
sample_size = 250
```
